#### Matplotlib Practice

The **plt** function allows us to quickly plot simple plots but we will need to use the **OOP Figure API** for more advanced things.

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.arange(0, 10)

y = 2 * x
```

We can see this is called the functional call since .plot is a function.

```python
# plot the relationship
plt.plot(x, y)
plt.show()  # hide output
```

```python
# add x, y labels and titles - has to be in the same cell
plt.plot(x, y)
plt.title('Plotting X and y')
plt.xlabel('x axis')
plt.ylabel('y axis')
```

```python
# adjusting x and y limits
plt.plot(x, y)
plt.xlim(0, 6)
plt.ylim(-2, 8)
# plt.savefit('my_simple_plot.png')  # provide full path if saving elsewhere
```
