#### More Advanced Plots wiht Figure Object

The OOP syntax uses the **Figure** object that is a blank figure. We then add axes to the figure and plot on the created axes.

+ plt.figure() creates a blank canvas for us to plot on
+ this has 0 axes
+ fig = plt.figure() is how we can access the blank canvas
+ axes = fig.add_axes([0, 0, 1, 1]) is how we can add one set of axes. The parameters are x, y, width and height. The x, y is the lower left corner. All numbers are ratios to the figure so 1, 1 takes up the entire figure space. 
+ since fig.add_axes([0, 0, 1, 1]) will take up the entire canvas starting at the lower left, if we say axes.plot(x, y) we will create a plot that takes up the entire space. Doing this is very similar to just calling plt.plot(x, y)

+ the power comes from creating multiple axes so that we can plot more than one graph on the figure.
+ most of the time we will just use plt.subplots() for this

When we make two axes with fractional numbers such as 0.2, 0.2, 0.5, 0.5 we are saying start 20% within the figure on the x and y axes and take up 50 percent of the width and 50 percent of the height.

```python
# create blank canvas
fig = plt.figure()

# large axes
axes1 = fig.add_axes([0, 0, 1, 1])  # same size as full canvas
axes1.plot(a, b);   # dont show output object type

# small axes
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.4])
```

### Figure Parameters

```python
# make a figure using figsize and dpi
# dpi is resolution - makes clearer
# figsize is length, width
fig = plt.figure(figsize = (11, 6), dpi = 200)

ax1 = fig.add_axes([0, 0, 1, 1])
ax1.plot(a, b)
ax1.set_title('Fourth Power')
ax1.set_xlim((0, 8))
ax1.set_ylim((0, 8000))
ax1.set_xlabel('x var')
ax1.set_ylabel(('y var'))
fig.savefig('figure.png', bbox_inches = 'tight')
```

Saving a figure using the OOP Figure API is a little different than .savefig() when using the functional .plot. When we save the figure we will have to adjust the size down and add bbox_inches = 'tight' so that we can see the axes.

### Subplots

**plt.subplots** makes creating figures in side by side formations for us so that we do not have to try to find a fig.add_axes([]) that works with the data we want. Subplots returns a tuple containing the Figure canvas and a numpy array holding the axes objects. This means we can index the axes when we say fig, axes = plt.subplots(). 

```python
# create data
a = np.linspace(0, 10, 11)
b = a ** 4
```

```python
x = np.arange(0, 10)
y = 2 * x
```

Here we input the nrows and ncols we want. We can plot the data above on the same canvas.

```python
# create subplots
fig, axes = plt.subplots(nrows = 1, ncols= 2)
```

With a row or column being 1 dimensional we can use loops to graph them.

```python
data = [
    np.random.randint(10, 101, size = 10),
    np.random.randint(101, 201, size = 10),
    np.random.randint(201, 301, size = 10)
]
```

```python
data
```

This is an example of how a for loop is commonly used when we have a 1-dimensional row or column. If we want to plot the three arrays within the data we created, we will enumerate the axes. As we see below, the first go in the for loop i = 1 and ax = axes[0] and so on and so fourth. So, when we say ax.plot(data[i]) in the for loop we first plot on ax[0] and we use data[0] as the data to plot.

```python
fig, axes = plt.subplots(nrows=3, ncols=1)

# Plot data on each subplot
for i, ax in enumerate(axes):
    # i = 0, ax = axes[0] (first subplot axis object)
    # i = 1, ax = axes[1] (second subplot axis object)
    # i = 2, ax = axes[2] (third subplot axis object)
    ax.plot(data[i])  # Plotting the corresponding data on each subplot
    ax.set_title(f'Subplot {i+1}')  # Setting the title of each subplot
```

```python
# fix the spacing
fig, axes = plt.subplots(nrows = 3, ncols = 1)

for i, ax in enumerate(axes):
    ax.plot(data[i])
    # add plus 1 since it starts at 0
    ax.set_title(f'{i + 1} plot')
    plt.tight_layout()
```

In this example we show how enumerate works over our data. When plotting though, we enumerate over the subplots and use the index for the data. This is because when we enumerate the data we will be looping through the data but we actually want to loop through the subplots and use the corresponding index as the data that will plotted.

```python
# refresher on enumerate
d = [[1, 3, 5], [7, 9, 11]]


for i, num in enumerate(d):
    print(i, num)
```

```python
fig, axes = plt.subplots(figsize = (8, 4), nrows=2, ncols=2)

# plot some data 
axes[0][0].plot(data[1])

# just use indexing for titles and such
axes[0][1].plot(x, y)
axes[0][1].set_title('linear')
axes[1][0].plot(data[0])
axes[1][1].plot(a, b)

# make sure no overlapping 
# these are fractional to axes so 1 width lets an entire 
# axes fit between
fig.subplots_adjust(wspace = 0.25, hspace = 0.2)

# figure level title
fig.suptitle('Figure-level Title')
```

##### Legends

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

# add labels that the legend will use
ax.plot(data[0], label = 'random data')
ax.plot(x, label = 'linear relationship')

# use loc for location - pair string or floats
ax.legend(loc = 'upper center')
```

We can also place it ourselves using the range when we add the axes - ([0, 0, 1, 1]). So to place it above the graph on the left hand side we use ax.legend(loc = (1, 1.1)) notice we just need two coordinates.

The possible linestyles are:
+ '--'
+ '-'
+ '-.'
+ ':'
+ 'steps'


```python
# linewidth
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.set_xlim(0, 10)
ax.set_ylim(0, 150)

# can do linewidth or lw
ax.plot(x, y, color = 'red', linewidth = '15', marker = '>', ms = 10.25, ls = '-.')
ax.plot(a, b, color = 'orange', lw = 5, marker = 'o', markersize = 20, linestyle = 'dotted')
ax.set_title('colors')
```
