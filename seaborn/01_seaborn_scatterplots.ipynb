### Seaborn

#### Scatterplots with Continuous Variables

Line up two continuous features as pairs and plot them as coordinates. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import data
df = pd.read_csv('../data/dm_office_sales.csv')
```

```python

# head
df.head()
```

Seaborn is built on top of matplotlib so we can edit the figure just as we did with matplotlib

```python
# use seaborn to plot scatterplot of sales and salary
sns.scatterplot(x = 'salary', y = 'sales', data = df)
```

```python
# figure size
plt.figure(figsize = (12, 4))
sns.scatterplot(x = 'salary', y = 'sales', data = df)
```

### Seaborn Parameters

Hue parameter takes in a column and this basically asks what column do you want to color the scatterplot points with. If we provide a categorical figure the legend is made for us based on the categories and if we do a continuous variable the graph will automatically select a gradient of that variable for us.

```python
plt.figure(figsize = (12, 4))
sns.scatterplot(x = 'salary', y = 'sales', hue = 'level of education',
                data = df)
```

##### Pallette

Use this link to select a color map. The **perceptually uniform sequential colormaps** are great for people who are color blind.

Most of the time we will be using a categorical variable for the hue. When we do this we will want to use one of the **qualitative colormaps*** from the link below.

https://matplotlib.org/stable/users/explain/colors/colormaps.html

We also introduce the **size** parameter here. The size parameter will size the points based on the variable you provide. So, in our example, when we provide 'salary' the points in the graph will be based on salary where the smaller points are for smaller salaries and the bigger points are for bigger salaries. This column can be continuous or categorical but will make more sense for continuous variables so keep that in mind.

```python
sns.scatterplot(x = 'salary', y = 'sales', 
                hue = 'level of education', palette= 'Set2',
                data = df, size = 'salary')
```

To make all the points bigger without depending on a variable we just use the s parameter.

```python
sns.scatterplot(x = 'salary', y ='sales',
            hue = 'level of education',
            palette = 'Dark2', s = 200, data = df)
```

To deal with points that are stacked on top of eachother we use the **alpha** parameter. This just makes the points transparent. If we set this to 0 everything will be completely see-through and we will see nothing.

```python
sns.scatterplot(x = 'salary', y = 'sales',
                hue = 'level of education',
                palette = 'Dark2', s = 150,
                alpha = 0.7, data = df)
```

Instead of using **hue** to color the points by a category, we can use the **style** parameter to plot them using different shapes based on the categorical column. In the example below, we use the **style** parameter on the level on education variable. We notice that we can also use *hue** in the same plot creation and each different shape will be a different color.

Using both **hue** and **style** does not show any additional information but it makes it easier to read. The style parameter is especially useful when we are publishing something in black and white.

```python
sns.scatterplot(x = 'salary',
                y = 'sales',
                s = 125, alpha = 0.7, hue = 'level of education',
                style = 'level of education',
                data = df)
```
