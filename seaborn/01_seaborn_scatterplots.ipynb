### Seaborn

#### Scatterplots with Continuous Variables

Line up two continuous features as pairs and plot them as coordinates. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import data
df = pd.read_csv('../data/dm_office_sales.csv')
```

```python

# head
df.head()
```

Seaborn is built on top of matplotlib so we can edit the figure just as we did with matplotlib

```python
# use seaborn to plot scatterplot of sales and salary
sns.scatterplot(x = 'salary', y = 'sales', data = df)
```

```python
# figure size
plt.figure(figsize = (12, 4))
sns.scatterplot(x = 'salary', y = 'sales', data = df)
```

### Seaborn Parameters

Hue parameter takes in a column and this basically asks what column do you want to color the scatterplot points with. If we provide a categorical figure the legend is made for us based on the categories and if we do a continuous variable the graph will automatically select a gradient of that variable for us.

```python
plt.figure(figsize = (12, 4))
sns.scatterplot(x = 'salary', y = 'sales', hue = 'level of education',
                data = df)
```
