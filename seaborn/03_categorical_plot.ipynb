## Categorical Plots

Categorical plots are similar to visualizing a groupby() call.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/dm_office_sales.csv')
```

```python
df.groupby('level of education')['salary'].agg(['mean', 'std'])
```

```python
plt.figure(figsize = (12, 6))
sns.boxplot(y = 'salary', x = 'level of education',
            data = df)
```
