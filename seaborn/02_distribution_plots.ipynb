## Distribution Plots

Distribution plots display a single continuous feature and help visualize things such as the mean and variance. The three main distribution plots are rug plot, histogram, and KDE.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

The rug plot just plots a tick for each value in the feature. This is not a very useful plot because the y-axis has no meaning and we cannot tell when ticks are stacked up on top of eachother. This plot is most useful for viewing outliers and the range of values.

The histogram is built directly off of the rug plot. To make a histogram from a rug plot we would count the amount of ticks in a given range such as salary between 60,000 and 70,000. The height of the graph is the count of ticks in that range.

The **KDE** plot is a Kernel Density Estimation plot on top of a histogram. It shows what a continuous probability function looks like on this data. This estimates a probability density function of the random variable we are dealing with. 

In our data we will not be working with any populations. For example, in the salary column, if we wanted to find the probability of someone having a 60,000 salary, we don't know this information because we are only dealing with a sample. The **KDE** plot allows us to deal with a continuous probability curve when we are really dealing with discrete measurements of our sample.

### Histogram

We can pass many of the matplotlib values to a seaborn plot as well such as color, linewidth....

```python
# set style will set the style for the entirenotebook
sns.set_style(style = 'darkgrid')
sns.displot(data = df, x = 'salary', kde = True,
            bins = 20,
            color = 'green', edgecolor = 'blue',
            linewidth = 4)
```
